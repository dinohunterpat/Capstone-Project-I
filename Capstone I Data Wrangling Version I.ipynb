```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Upload China data
China_df = pd.read_csv('China Meta Data.csv')
```

```python
China_df.head()
```

```python
list_labels_China = ['Year','Tickets Sold (1000)','Total Gross (CNY)','Total Gross (USD)', 'Number of Screens',
                     'Ticket Price (CNY)','Ticket Price (USD)','Movies Released','Dollar Purchasing Parity',
                     'Dollar Valuation (%)','Population','Average Wage (CNY)','Average Wage (USD)']
```

```python
# Create new labels for China data
China_df.columns = list_labels_China
China_df.head()
```

```python
# Eliminate Commas
China_df['Tickets Sold (1000)'] = China_df['Tickets Sold (1000)'].replace({',': ''}, regex=True)
China_df['Total Gross (CNY)'] = China_df['Total Gross (CNY)'].replace({',': ''}, regex=True)
China_df['Number of Screens'] = China_df['Number of Screens'].replace({',': ''}, regex=True)
China_df.head()
```

```python
# Convert values to numeric
col = China_df.columns
China_df[col] = China_df[col].apply(pd.to_numeric, errors='coerce')
China_df=China_df.dropna()
China_df
```

```python
# Upload Japan data
Japan_df = pd.read_csv('Japan_Meta_Modified.csv')
Japan_df.head()
```

```python
list_labels_Japan = ['Year','Tickets Sold (1000)','Total Gross (JPY)','Total Gross (USD)', 'Number of Screens',
                     'Ticket Price (JPY)','Ticket Price (USD)','Movies Released','Dollar Purchasing Parity',
                     'Dollar Valuation (%)','Population','Average Wage (JPY)','Average Wage (USD)']
```

```python
# Create new labels
Japan_df.columns = list_labels_Japan
Japan_df.head()
```

```python
# Remove commas
Japan_df['Tickets Sold (1000)'] = Japan_df['Tickets Sold (1000)'].replace({',': ''}, regex=True)
Japan_df['Total Gross (JPY)'] = Japan_df['Total Gross (JPY)'].replace({',': ''}, regex=True)
Japan_df['Number of Screens'] = Japan_df['Number of Screens'].replace({',': ''}, regex=True)
Japan_df.head()
```

```python
# Convert values to numeric
col = Japan_df.columns
Japan_df[col] = Japan_df[col].apply(pd.to_numeric, errors='coerce')
Japan_df
```

```python
#result = pd.concat([df1, df4], axis=1, join_axes=[df1.index])
Total = pd.concat([China_df,Japan_df], axis=1, ignore_index = True)
Total
```

```python
list_labels_Total = ['Year','Tickets Sold in China (1000)','Total Gross in China (CNY)','Total Gross in China (USD)',
                     'Number of Screens in China', 'Ticket Price in China (CNY)','Ticket Price in China (USD)',
                     'Movies Released in China','Dollar Purchasing Parity in China',
                     'Dollar Valuation in China (%)','Population in China','Average Wage in China (CNY)',
                     'Average Wage in China (USD)','Year1','Tickets Sold in Japan (1000)','Total Gross in Japan (JPY)',
                     'Total Gross in Japan (USD)', 'Number of Screens in Japan','Ticket Price in Japan (JPY)',
                     'Ticket Price in Japan (USD)','Movies Released in Japan','Dollar Purchasing Parity in Japan',
                     'Dollar Valuation in Japan (%)','Population in Japan','Average Wage in Japan (JPY)','Average Wage in Japan (USD)']
Total.columns = list_labels_Total
Total = Total.drop(columns=['Year1'])
Total.head()
#China_salary_adj = (df['Ticket Price (CNY)'] * 
#result['Adjusted Salary in China'] = full_price
#result['Adjusted Ticket Price in China']
```

```python
# Total Gross
Total.plot(x='Year', y=['Total Gross in China (USD)','Total Gross in Japan (USD)'])
plt.title('Total Gross')
plt.xlabel('Year')
plt.ylabel('Box Office Gross in Billions USD')
plt.show()
```

```python
# Tickets Sold
Total.plot(x='Year', y=['Tickets Sold in China (1000)','Tickets Sold in Japan (1000)'],logy=True)
plt.title('Tickets Sales')
plt.xlabel('Year')
plt.ylabel('Tickets Sold in Thousands')
plt.show()
```

```python
# Ticket Prices
Total.plot(x='Year', y=['Ticket Price in Japan (USD)','Ticket Price in China (USD)'])
plt.title('Ticket Prices')
plt.xlabel('Year')
plt.ylabel('Ticket Price in USD')
plt.show()
```

```python
# Number of Screens
Total.plot(x='Year', y=['Number of Screens in China','Number of Screens in Japan'])
plt.title('Number of Screens')
plt.xlabel('Years')
plt.ylabel('Number of Screens')
plt.show()
```

```python
# Movies Released
Total.plot(x='Year', y=['Movies Released in China','Movies Released in Japan'])
plt.title('Numbers of New Releases')
plt.xlabel('Years')
plt.ylabel('New Movies')
plt.show()
```

```python
# Income Growth
Total.plot(x='Year', y=['Average Wage in China (USD)','Average Wage in Japan (USD)'],logy=True)
plt.title('Average Wage over Time')
plt.xlabel('Years')
plt.ylabel('Average Wage in USD')
plt.show()
```

```python
# Population Growth
Total.plot(x='Year', y=['Population in China','Population in Japan'])
plt.title('Population Growth')
plt.xlabel('Years')
plt.ylabel('Population in Billions')
plt.show()
```

```python
# Correlation between ticket price and gross
China_Gross_Ticket = Total[["Total Gross in China (USD)","Ticket Price in China (USD)"]]
print(China_Gross_Ticket.corr())
Japan_Gross_Ticket = Total[["Total Gross in Japan (USD)","Ticket Price in Japan (USD)"]] 
print(Japan_Gross_Ticket.corr())
```
