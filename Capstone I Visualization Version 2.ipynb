```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Upload China data
China_df = pd.read_csv('China Meta Data Modified.csv')
China_df.head()
```

```python
# Create new labels for China data
list_labels_China = ['Year','Tickets Sold (Millions)','Total Gross (Millions CNY)','Total Gross (Millions USD)', 'Number of Screens',
                     'Ticket Price (CNY)','Ticket Price (USD)','Movies Released','Dollar Purchasing Parity',
                     'Dollar Valuation (%)','Population','Average Wage (CNY)','Average Wage (USD)','USD:CNY Exchange Rate',
                     'Tickets Sold per 100,000 Population','Gross per Capita (USD)']
China_df.columns = list_labels_China
China_df.head()
```

```python
# Eliminate Commas
China_df['Tickets Sold (Millions)'] = China_df['Tickets Sold (Millions)'].replace({',': ''}, regex=True)
China_df['Number of Screens'] = China_df['Number of Screens'].replace({',': ''}, regex=True)
China_df['Tickets Sold per 100,000 Population'] = China_df['Tickets Sold per 100,000 Population'].replace({',': ''}, regex=True)
China_df.head()
```

```python
# Convert values to numeric
col = China_df.columns
China_df[col] = China_df[col].apply(pd.to_numeric, errors='coerce')
China_df=China_df.dropna()
China_df.head()
```

```python
# Upload Japan data
Japan_df = pd.read_csv('Japan Meta Data Modified.csv')
Japan_df.head()
```

```python
list_labels_Japan = ['Year','Tickets Sold (Millions)','Total Gross (Millions JPY)','Total Gross (Millions USD)', 
                     'Number of Screens', 'Ticket Price (JPY)','Ticket Price (USD)','Movies Released',
                     'Dollar Purchasing Parity','Dollar Valuation (%)','Population','Average Wage (JPY)',
                     'Average Wage (USD)','USD to JPY Exchange Rate','Tickets Sold per 100,000 Population','Gross per Capita (USD)']
```

```python
# Create new labels
Japan_df.columns = list_labels_Japan
Japan_df.head()
```

```python
# Remove commas
Japan_df['Total Gross (Millions JPY)'] = Japan_df['Total Gross (Millions JPY)'].replace({',': ''}, regex=True)
Japan_df['Number of Screens'] = Japan_df['Number of Screens'].replace({',': ''}, regex=True)
```

```python
# Convert values to numeric
col = Japan_df.columns
Japan_df[col] = Japan_df[col].apply(pd.to_numeric, errors='coerce')
Japan_df.head()
```

```python
#result = pd.concat([df1, df4], axis=1, join_axes=[df1.index])
Total = pd.concat([China_df,Japan_df], axis=1, ignore_index = True)
Total.head()
```

```python
list_labels_Total = ['Year','Tickets Sold in China (Millions)','Total Gross in China (Millions CNY)','Total Gross in China (Millions USD)',
                     'Number of Screens in China',
                     'Ticket Price in China (CNY)','Ticket Price in China (USD)','Movies Released in China','Dollar Purchasing Parity in China',
                     'Dollar Valuation in China (%)','Population in China','Average Wage in China (CNY)','Average Wage in China (USD)',
                     'USD:CNY Exchange Rate',
                     'Tickets Sold per 100,000 Population in China','Gross per Capita in China (USD)','Year1','Tickets Sold in Japan (Millions)',
                     'Total Gross in Japan (Millions JPY)','Total Gross in Japan (Millions USD)', 
                     'Number of Screens in Japan', 'Ticket Price in Japan (JPY)','Ticket Price in Japan (USD)','Movies Released in Japan',
                     'Dollar Purchasing Parity in Japan','Dollar Valuation in Japan (%)','Population in Japan','Average Wage in Japan (JPY)',
                     'Average Wage in Japan (USD)','USD to JPY Exchange Rate','Tickets Sold per 100,000 Population in Japan','Gross per Capita in Japan (USD)'] 
Total.columns = list_labels_Total
Modified_Total = Total.drop(columns=['Year1'])
Modified_Total.head()
```

```python
# Total Gross (Line)
Modified_Total.plot(x='Year', y=['Total Gross in China (Millions USD)','Total Gross in Japan (Millions USD)'])
plt.title('Total Gross')
plt.xlabel('Year')
plt.ylabel('Box Office Gross in Millions USD')
plt.show()
```

```python
# Gross Per Capita (Line)
Modified_Total.plot(x='Year', y=['Gross per Capita in China (USD)','Gross per Capita in Japan (USD)'])
plt.title('Gross Per Capita')
plt.xlabel('Year')
plt.ylabel('Gross per Capita (USD)')
plt.show()
```

```python
# Tickets Sold (Line)
Modified_Total.plot(x='Year', y=['Tickets Sold in China (Millions)','Tickets Sold in Japan (Millions)'])
plt.title('Tickets Sold')
plt.xlabel('Year')
plt.ylabel('Tickets Sold in Millions')
plt.show()
```

```python
# Tickets Solds Per Hundred Thousand (line) 
Modified_Total.plot(x='Year', y=['Tickets Sold per 100,000 Population in China',
                        'Tickets Sold per 100,000 Population in Japan'])
plt.title('Tickets Sold Per Hundred Thousand')
plt.xlabel('Year')
plt.ylabel('Tickets Sold Per Hundred Thousand People')
plt.show()
```

```python
# Ticket Prices (Line)
Modified_Total.plot(x='Year', y=['Ticket Price in Japan (USD)','Ticket Price in China (USD)'])
plt.title('Ticket Prices')
plt.xlabel('Year')
plt.ylabel('Ticket Price in USD')
plt.legend(loc='center right')
plt.show()
```

```python
# Number of Screens (Line)
Modified_Total.plot(x='Year', y=['Number of Screens in China','Number of Screens in Japan'])
plt.title('Number of Screens')
plt.xlabel('Years')
plt.legend(loc='upper left')
plt.ylabel('Number of Screens')
plt.show()
```

```python
# Movies Released (Line)
Modified_Total.plot(x='Year', y=['Movies Released in China','Movies Released in Japan'])
plt.title('Numbers of New Releases')
plt.xlabel('Years')
plt.ylabel('New Movies')
plt.legend(loc='center right')
plt.show()
```

```python
# Income Growth (Line)
Modified_Total.plot(x='Year', y=['Average Wage in China (USD)','Average Wage in Japan (USD)'])
plt.title('Average Wage over Time')
plt.xlabel('Years')
plt.ylabel('Average Wage in USD')
plt.show()
```

```python
# Number of Screens (Bar Graph)
index = np.arange(5)
bar_width = 0.35

fig, ax = plt.subplots()
screens_china=ax.bar(index,Modified_Total['Number of Screens in China'],
                     bar_width, label="China")
screens_japan = ax.bar(index+bar_width, Modified_Total["Number of Screens in Japan"],
                 bar_width, label="Japan")
ax.set_xlabel('Year')
ax.set_ylabel('Screens')
ax.set_title('Number of Screens: China vs Japan')
ax.set_xticks(index + bar_width / 2)
ax.set_xticklabels(["2012", "2013", "2014", "2015", "2016"])
ax.legend()
plt.show()
```

```python
# Tickets Sold (Bar Graph)
fig, ax = plt.subplots()
screens_china=ax.bar(index,Modified_Total['Tickets Sold in China (Millions)'],
                     bar_width, label="China")
screens_japan = ax.bar(index+bar_width, Modified_Total['Tickets Sold in Japan (Millions)'],
                 bar_width, label="Japan")
ax.set_xlabel('Year')
ax.set_ylabel('Tickets Sold (Millions)')
ax.set_title('Tickets Sold: China vs Japan')
ax.set_xticks(index + bar_width / 2)
ax.set_xticklabels(["2012", "2013", "2014", "2015", "2016"])
ax.legend()
plt.show()
```

```python
# Tickets Sold per 100,000 (Bar Graph)
fig, ax = plt.subplots()
screens_china=ax.bar(index,Modified_Total['Tickets Sold per 100,000 Population in China'],
                     bar_width, label="China")
screens_japan = ax.bar(index+bar_width, Modified_Total['Tickets Sold per 100,000 Population in Japan'],
                 bar_width, label="Japan")
ax.set_xlabel('Year')
ax.set_ylabel('Tickets Sold per 100,000 People')
ax.set_title('Tickets Sold per 100,000: China vs Japan')
ax.set_xticks(index + bar_width / 2)
ax.set_xticklabels(["2012", "2013", "2014", "2015", "2016"])
ax.legend()
plt.show()
```

```python
# Total Gross (Bar Graph)
fig, ax = plt.subplots()
screens_china=ax.bar(index,Total['Total Gross in China (Millions USD)'],
                     bar_width, label="China")
screens_japan = ax.bar(index+bar_width, Total["Total Gross in Japan (Millions USD)"],
                 bar_width, label="Japan")
ax.set_xlabel('Year')
ax.set_ylabel('Total Gross in USD')
ax.set_title('Total Gross: China vs Japan')
ax.set_xticks(index + bar_width / 2)
ax.set_xticklabels(["2012", "2013", "2014", "2015", "2016"])
ax.legend()
plt.show()
```

```python
# Total Gross per Capita (Bar Graph)
fig, ax = plt.subplots()
screens_china=ax.bar(index,Total['Gross per Capita in China (USD)'],
                     bar_width, label="China")
screens_japan = ax.bar(index+bar_width, Total["Gross per Capita in Japan (USD)"],
                 bar_width, label="Japan")
ax.set_xlabel('Year')
ax.set_ylabel('Gross per Capita in USD')
ax.set_title('Gross per Capita: China vs Japan')
ax.set_xticks(index + bar_width / 2)
ax.set_xticklabels(["2012", "2013", "2014", "2015", "2016"])
ax.legend()
plt.show()
```
